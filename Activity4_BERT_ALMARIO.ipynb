# Activity 4: Understanding Context with BERT

# Step 1: Install and Import Required Libraries

```python
# Install transformers (if not installed)
!pip install transformers

# Import the pipeline function from transformers
from transformers import pipeline
```

# Step 2: Load the BERT Model for Word Sense Disambiguation

```python
# Load a BERT model for predicting masked words
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
```

# Step 3: Define Sentences and Use BERT to Predict Word Meaning

```python
# Example sentences
sentence1 = "The bank is next to the river."
sentence2 = "She deposited her paycheck in the bank."

# Replace "bank" with [MASK] for BERT to predict the correct word
masked_sentence1 = sentence1.replace("bank", "[MASK]")
masked_sentence2 = sentence2.replace("bank", "[MASK]")

# Print predictions
print("\nContext 1:", sentence1)
print("Prediction:", fill_mask(masked_sentence1))

print("\nContext 2:", sentence2)
print("Prediction:", fill_mask(masked_sentence2))
```
